#Criando gr√°ficos com massa de dados em CSV
Importando dados do GitHub

```python
import pandas as pd

# Pegando os dados diretamente do git
url = 'https://raw.githubusercontent.com/Red-of-Pallet/Estudo/dataScience/dataScience/populacao_brasileira/populacao_brasileira.csv'

df1 = pd.read_csv(url, delimiter=';')
df1.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Titulos
plt.title('Crescimento da Popula√ß√£o Brasileira 1980-2016')
plt.xlabel('Ano')
plt.ylabel('Popula√ß√£o x100.000.000')

# Dados
plt.plot(df1['ano'], df1['population'], color='k', linestyle='--')
plt.bar(df1['ano'], df1['population'], color='#D3D3D3')

plt.show()
```
