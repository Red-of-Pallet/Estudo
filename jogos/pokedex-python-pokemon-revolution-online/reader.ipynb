```python
import json

def combinar_jsons(json1, json2):
    resultado = []

    for entrada in json1:
        entrada_combinada = {
            'Regiao': '',
            'Mapa': entrada['Map'],
            'Pokemon': entrada['Pokemon'],
            'Type': '',
            'MonsterID': entrada['MonsterID'],
            'Daytime': entrada['Daytime'],
            'MinLVL': entrada['MinLVL'],
            'MaxLVL': entrada['MaxLVL'],
            'Item': entrada['Item'],
            'MemberOnly': entrada['MemberOnly'],
            'Tier': entrada['Tier'],
            'Fishing': -1,        # Valor padrão para coluna ausente
            'FishingOnly': -1,    # Valor padrão para coluna ausente
            'RequiredRod': '',     # Valor padrão para coluna ausente
            'HP': '',
            'Attack': '',
            'Defense': '',
            'Sp. Attack': '',
            'Sp. Defense': '',
            'Speed': ''
        }
        resultado.append(entrada_combinada)

    for entrada in json2:
        entrada_combinada = {
            'Regiao': '',
            'Mapa': entrada['Map'],
            'Pokemon': entrada['Pokemon'],
            'Type': '',
            'MonsterID': entrada['MonsterID'],
            'Daytime': entrada['Daytime'],
            'MinLVL': entrada['MinLVL'],
            'MaxLVL': entrada['MaxLVL'],
            'Item': entrada['Item'],
            'MemberOnly': entrada['MemberOnly'],
            'Tier': entrada['Tier'],
            'Fishing': entrada['Fishing'],
            'FishingOnly': entrada['FishingOnly'],
            'RequiredRod': entrada['RequiredRod'],
            'HP': '',
            'Attack': '',
            'Defense': '',
            'Sp. Attack': '',
            'Sp. Defense': '',
            'Speed': ''
        }
        resultado.append(entrada_combinada)

    return resultado
```

```python
def preencherAreas(base_data, areas_data, initial_data, area):
    for j in areas_data:
        for k in j['Areas']:
            for i in base_data:
                if(i['Mapa'] == k):
                    entrada = {
                        'Regiao': area,
                        'Mapa': i['Mapa'],
                        'Pokemon': i['Pokemon'],
                        'Type': '',
                        'MonsterID': i['MonsterID'],
                        'Daytime': i['Daytime'],
                        'MinLVL': i['MinLVL'],
                        'MaxLVL': i['MaxLVL'],
                        'Item': i['Item'],
                        'MemberOnly': i['MemberOnly'],
                        'Tier': i['Tier'],
                        'Fishing': i['Fishing'],
                        'FishingOnly': i['FishingOnly'],
                        'RequiredRod': i['RequiredRod'],
                        'HP': '',
                        'Attack': '',
                        'Defense': '',
                        'Sp. Attack': '',
                        'Sp. Defense': '',
                        'Speed': ''
                    }
                    initial_data.append(entrada)

    base_data = [i for i in base_data if i not in initial_data]

    return base_data, initial_data
```

```python

def preencherStatus(initial_data, pokedex_data):
    exibir = 0

    for entry in pokedex_data:
        for i in initial_data:
            if entry['id'] == i['MonsterID']:
                i.update({
                    'Type': entry.get('type', ''),
                    'HP': entry.get('HP', ''),
                    'Attack': entry.get('Attack', ''),
                    'Defense': entry.get('Defense', ''),
                    'Sp. Attack': entry.get('Sp. Attack', ''),
                    'Sp. Defense': entry.get('Sp. Defense', ''),
                    'Speed': entry.get('Speed', ''),
                })
                if(exibir % 100 == 0):
                    print(i)
                exibir += 1

    return initial_data
```

```python
def salvar_para_json(dados, nome_arquivo):
    with open(nome_arquivo, 'w') as f:
        json.dump(dados, f, indent=2)
```

```python

f = open("json/base/land.json")
land_data = json.load(f)

f = open('json/base/surf.json')
surf_data = json.load(f)

f = open('json/base/maps.json')
regions_data = json.load(f)

with open('json/pokedex.json', encoding='utf-8') as f:
    pokedex_data = json.load(f)

# display(pokedex_data)
```
